```python
#seleksi fitur
```

```python
features = ['year','mileage','tax','mpg','engineSize']
x = df[features]
y = df['price']
x.shape, y.shape
```

#split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

#membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
from sklearn.linear_model import LinearRegression
Ir = LinearRegression()
Ir.fit(x_train,y_train)
pred = Ir.predict(X_test)
```

```python
score = Ir.score(X_test, y_test)
print('akurasi model regresi lidier = ', score)
```

#membuat input model regresi linier

```python
#Year=2019, mileage=5000, tax=145, mpg=30.2, engineSize=2
input_data = np.array([[2019, 5000, 145, 30.2,2]])
prediction = Ir.predict(input_data)
print('Estimasi harga mobil dalam EUR :', prediction)
```
